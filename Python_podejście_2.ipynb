```python
# Podstawowe operacje matematyczne w Pythonie

# Dodawanie
a = 5
b = 3
suma = a + b
print(f"Dodawanie: {a} + {b} = {suma}")

# Odejmowanie
roznica = a - b
print(f"Odejmowanie: {a} - {b} = {roznica}")

# Mnożenie
iloczyn = a * b
print(f"Mnożenie: {a} * {b} = {iloczyn}")

# Dzielenie
dzielenie = a / b
print(f"Dzielenie: {a} / {b} = {dzielenie}")

# Dzielenie całkowite (bez reszty)
dzielenie_calkowite = a // b
print(f"Dzielenie całkowite: {a} // {b} = {dzielenie_calkowite}")

# Reszta z dzielenia (modulo)
reszta = a % b
print(f"Reszta z dzielenia: {a} % {b} = {reszta}")

# Potęgowanie
potega = a ** b
print(f"Potęgowanie: {a} ** {b} = {potega}")

# Przykładowe operacje z użyciem zmiennych od użytkownika
c = int(input("Podaj pierwszą liczbę: "))
d = int(input("Podaj drugą liczbę: "))
print(f"Dodawanie: {c} + {d} = {c + d}")
print(f"Mnożenie: {c} * {d} = {c * d}")

# Potęgowanie z obsługą dużych liczb
try:
    potega_user = c ** d
    print(f"Potęgowanie: {c} ** {d} = {potega_user}")
except ValueError as e:
    print(f"Błąd podczas potęgowania: {e}")
```

```python
# Podstawowe operacje matematyczne w Pythonie

# Dodawanie
a = 5
b = 3
suma = a + b
print(f"Dodawanie: {a} + {b} = {suma}")

# Odejmowanie
roznica = a - b
print(f"Odejmowanie: {a} - {b} = {roznica}")

# Mnożenie
iloczyn = a * b
print(f"Mnożenie: {a} * {b} = {iloczyn}")

# Dzielenie
dzielenie = a / b
print(f"Dzielenie: {a} / {b} = {dzielenie}")

# Dzielenie całkowite (bez reszty)
dzielenie_calkowite = a // b
print(f"Dzielenie całkowite: {a} // {b} = {dzielenie_calkowite}")

# Reszta z dzielenia (modulo)
reszta = a % b
print(f"Reszta z dzielenia: {a} % {b} = {reszta}")

# Potęgowanie
potega = a ** b
print(f"Potęgowanie: {a} ** {b} = {potega}")

# Przykładowe operacje z użyciem zmiennych od użytkownika
c = int(input("Podaj pierwszą liczbę: "))
d = int(input("Podaj drugą liczbę: "))
print(f"Dodawanie: {c} + {d} = {c + d}")
print(f"Mnożenie: {c} * {d} = {c * d}")

# Potęgowanie z obsługą dużych liczb
try:
    potega_user = c ** d
    print(f"Potęgowanie: {c} ** {d} = {potega_user}")
except ValueError as e:
    print(f"Błąd podczas potęgowania: {e}")

# Operacje na listach
lista = [1, 2, 3, 4, 5]
print(f"Lista: {lista}")
lista.append(6)  # Dodawanie elementu do listy
print(f"Lista po dodaniu elementu: {lista}")
lista.remove(3)  # Usuwanie elementu z listy
print(f"Lista po usunięciu elementu: {lista}")
print(f"Drugi element listy: {lista[1]}")  # Dostęp do elementu listy

# Operacje na słownikach
slownik = {"jeden": 1, "dwa": 2, "trzy": 3}
print(f"Słownik: {slownik}")
slownik["cztery"] = 4  # Dodawanie nowej pary klucz-wartość
print(f"Słownik po dodaniu pary: {slownik}")
slownik["dwa"] = 22  # Aktualizacja wartości dla istniejącego klucza
print(f"Słownik po aktualizacji wartości: {slownik}")

# Operacje na krotkach
krotka = (10, 20, 30, 40)
print(f"Krotka: {krotka}")
print(f"Pierwszy element krotki: {krotka[0]}")
# krotka[1] = 25  # To spowoduje błąd, ponieważ krotki są niemodyfikowalne
```

# Pandas

```python
# Analiza danych z użyciem Pandas

import pandas as pd

# Wczytywanie danych z pliku CSV do DataFrame
# Zakładamy, że mamy plik 'dane.csv' w Google Colab (możesz przesłać plik bezpośrednio do Colab)
df = pd.read_csv('dane.csv')

# Wyświetlenie pierwszych kilku wierszy danych
print("Pierwsze 5 wierszy danych:")
print(df.head())

# Podstawowe informacje o danych
print("\nPodstawowe informacje o danych:")
print(df.info())

# Statystyki opisowe dla danych liczbowych
print("\nStatystyki opisowe:")
print(df.describe())

# Sprawdzenie brakujących wartości
df_missing = df.isnull().sum()
print("\nLiczba brakujących wartości w każdej kolumnie:")
print(df_missing)

# Filtrowanie danych - przykładowo: wybieramy wiersze, w których wartość w kolumnie 'wiek' jest większa niż 30
wiek_powyzej_30 = df[df['wiek'] > 30]
print("\nWiersze, w których wiek jest większy niż 30:")
print(wiek_powyzej_30)

# Grupowanie danych - grupowanie po kolumnie 'miasto' i obliczenie średniej wartości w kolumnie 'dochód'
miasto_do_sredniego_dochodu = df.groupby('miasto')['dochód'].mean()
print("\nŚredni dochód w zależności od miasta:")
print(miasto_do_sredniego_dochodu)

# Dodanie nowej kolumny - obliczanie kolumny 'wiek_plus_5' dodającej 5 lat do wartości z kolumny 'wiek'
df['wiek_plus_5'] = df['wiek'] + 5
print("\nDane po dodaniu nowej kolumny 'wiek_plus_5':")
print(df.head())

# Zapis zaktualizowanych danych do nowego pliku CSV
df.to_csv('dane_zaktualizowane.csv', index=False)
print("\nZaktualizowane dane zostały zapisane do pliku 'dane_zaktualizowane.csv'")
```

```python
# Tworzenie przykładowego pliku CSV z danymi
import pandas as pd

# Definiowanie przykładowych danych
dane = {
    'imie': ['Anna', 'Piotr', 'Kasia', 'Marek', 'Julia'],
    'wiek': [28, 35, 22, 40, 30],
    'miasto': ['Warszawa', 'Kraków', 'Poznań', 'Gdańsk', 'Wrocław'],
    'dochód': [4000, 5000, 3200, 5800, 4300]
}

# Tworzenie DataFrame z danych
df = pd.DataFrame(dane)

# Zapisanie danych do pliku CSV
df.to_csv('dane.csv', index=False)

print("Przykładowy plik 'dane.csv' został utworzony.")
```

# Wizualizacja danych

```python
# Wizualizacja danych z użyciem Matplotlib i Seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Wczytanie danych z pliku CSV
df = pd.read_csv('dane.csv')

# Wyświetlenie danych
print("Dane wczytane z pliku CSV:")
print(df.head())

# Wizualizacja rozkładu wieku
plt.figure(figsize=(8, 5))
sns.histplot(df['wiek'], bins=5, kde=True, color='blue')
plt.xlabel('Wiek')
plt.ylabel('Liczba osób')
plt.title('Rozkład wieku wśród osób')
plt.show()

# Wizualizacja średniego dochodu w zależności od miasta
plt.figure(figsize=(8, 5))
sns.barplot(x='miasto', y='dochód', data=df, palette='viridis')
plt.xlabel('Miasto')
plt.ylabel('Średni dochód')
plt.title('Średni dochód w zależności od miasta')
plt.show()

# Wizualizacja zależności między wiekiem a dochodem
plt.figure(figsize=(8, 5))
sns.scatterplot(x='wiek', y='dochód', data=df, hue='miasto', palette='deep', s=100)
plt.xlabel('Wiek')
plt.ylabel('Dochód')
plt.title('Zależność między wiekiem a dochodem')
plt.legend(title='Miasto', loc='best')
plt.show()

# Wizualizacja typu pudełkowego (boxplot) dochodów w różnych miastach
plt.figure(figsize=(8, 5))
sns.boxplot(x='miasto', y='dochód', data=df, palette='Set2')
plt.xlabel('Miasto')
plt.ylabel('Dochód')
plt.title('Porównanie dochodów w różnych miastach')
plt.show()
```
